```python
# what about creating a smaller DataFrame, just bill length vs species?
# a subset ...
# on selecting data: https://jakevdp.github.io/PythonDataScienceHandbook/03.02-data-indexing-and-selection.html

bill_Length_species = penguins[["Bill Length (mm)", "Species"]]

# let's sort that again
bill_Length_species.sort_values("Bill Length (mm)", ascending =False).head(10)
```

```python
# if we are going to visualize this, we could choose to get rid of NA outliers

bill_Length_species = bill_Length_species.dropna()
bill_Length_species.shape

# more context on missing values: https://jakevdp.github.io/PythonDataScienceHandbook/03.04-missing-values.html
```

```python
# let's look @ the mean of the bill length, by species
# documentation on groupby: https://pandas.pydata.org/docs/user_guide/groupby.html

bill_Length_species.groupby("Species").mean()
```

```python
from pandas.errors import InvalidIndexError
# let's start with a scatter plot of just this data, species + bill length
# working with the matplotlib defaults ...

# we need to load matplotlib!

import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(bill_Length_species["Species"], bill_Length_species["Bill Length (mm)"])
```

```python
# but if I want to show each species in a different color, I should filter my data

# make a list of all the unique values in the "Species" column, full names
species = bill_Length_species["Species"].unique()
print(species)

# filter: just DataFrames of culmen length vs. species, one DataFrame per species
adelie = bill_Length_species[bill_Length_species["Species"]==species[0]]

# note we could loop here ... but are spelling this out for now
gentoo = bill_Length_species[bill_Length_species["Species"]==species[1]]

chinstrap = bill_Length_species[bill_Length_species["Species"]==species[2]]



# let's check on just chinstrap, with head()
chinstrap.head()
```

```python
# now we can plot the 3 individually, with different options!
# remember: plt.scatter(x, y, options)
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

#alpha
#In this example, the alpha value of 0.5 makes the points in the scatter plot semi-transparent.
#Using alpha can be particularly helpful when you have multiple data series or elements in a plot,
#and you want to make it easier to distinguish overlapping or densely packed points or elements.
#It's a way to add a level of visual depth to your plots.
#s parameter is used to specify the size of markers (points) in a scatter plot.


plt.scatter(adelie["Species"], adelie["Bill Length (mm)"], color ="Orange", alpha =0.5, s=150)
plt.scatter(gentoo["Species"], gentoo["Bill Length (mm)"], color ="Red", alpha =0.5, s=150)
plt.scatter(chinstrap["Species"], chinstrap["Bill Length (mm)"], color ="Green", alpha =0.5, s= 150)
```

```python
# to examine the mean along with the scatter, let's add a boxplot to each species
# matplotlib docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html#

y_data = [adelie["Bill Length (mm)"], gentoo["Bill Length (mm)"], chinstrap["Bill Length (mm)"]]

# remember the pattern! plt.type(data, options)
plt.boxplot(y_data, positions=[0,1,2])

# now we add the scatters back in
# pattern: plt.scatter(x, y, options)
# to match the box plots to the scatter plots, we assign:
# adelie = 0, gentoo = 1, chinstrap = 2
# x values here need to be a list the same length as the y list


plt.scatter([0] * len(adelie), adelie["Bill Length (mm)"], color = "Orange", alpha =0.5,s=100)
plt.scatter([1] * len(gentoo), gentoo["Bill Length (mm)"], color = "Red", alpha =0.5, s=100 )
plt.scatter([2] * len(chinstrap), chinstrap["Bill Length (mm)"], color = "Green", alpha =0.5, s=100)
```

```python
# let's do some data transformation, and "jitter" the dots

plt.boxplot(y_data, positions = [0,1,2])

import random

# for each species, I want to have an x_data list, small random numbers in a range near their x value
# the list of x values needs to be the same length as the list of y values

x_adelie = [random.uniform(-0.1, 0.1) + 0 for numbers in range (len(adelie))]
print(x_adelie)

x_gentoo = [random.uniform(-0.1, 0.1) + 1 for numbers in range (len(gentoo))]
print(gentoo)

x_chinstrap = [random.uniform(-0.1, 0.1) + 2 for numbers in range (len(chinstrap))]
print(chinstrap)

plt.scatter(x_adelie, adelie["Bill Length (mm)"], color = "Orange", alpha =0.5, s=100)
plt.scatter(x_gentoo, gentoo["Bill Length (mm)"], color = "Red", alpha =0.5, s=100)
plt.scatter(x_chinstrap, chinstrap["Bill Length (mm)"], color = "Green", alpha =0.5, s=100)
```

```python
# let's see a violin plot to this data
# documentation: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.violinplot.html

violins = plt.violinplot(y_data, positions = [0,1,2], widths = 0.4, bw_method="silverman")
```

```python
# now we ADD the violin to the outside

# and style the median lines in the box plot

plt.boxplot(y_data, positions= [0,1,2], medianprops = {"color": "Purple", "linewidth": 2})

plt.scatter(x_adelie, adelie["Bill Length (mm)"], color = "Orange", alpha =0.5, s=100)
plt.scatter(x_gentoo, gentoo["Bill Length (mm)"], color = "Red", alpha =0.5, s=100)
plt.scatter(x_chinstrap, chinstrap["Bill Length (mm)"], color = "Green", alpha =0.5, s=100)

vilions = plt.violinplot(y_data, positions =[0,1,2], widths =0.4, bw_method = "silverman")

# let's change the background fill to transparent - or, facecolor to 'none' for each violin plot
for body in violins['bodies']:
  body.set_facecolor('none')
  body.set_edgecolor('Black')
```

```python
# reproducing the plot ...
plt.figure(figsize=(14, 10))
# figsize = inches!

plt.boxplot(y_data, positions= [0,1,2], medianprops = {"color": "Purple", "linewidth": 2})

plt.scatter(x_adelie, adelie["Bill Length (mm)"], color = "Orange", alpha =0.5, s=100)
plt.scatter(x_gentoo, gentoo["Bill Length (mm)"], color = "Red", alpha =0.5, s=100)
plt.scatter(x_chinstrap, chinstrap["Bill Length (mm)"], color = "Green", alpha =0.5, s=100)

vilions = plt.violinplot(y_data, positions =[0,1,2], widths =0.4, bw_method = "silverman")

# let's change the background fill to transparent - or, facecolor to 'none' for each violin plot
for body in violins['bodies']:
  body.set_facecolor('none')
  body.set_edgecolor('Black')


# but let's clean up the axes, titles, text, font, etc.!
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html

plt.rcParams["font.family"]= "monospace"
plt.xticks([0,1,2], ["Adelie", "Gentoo", "Chinstrap"])
plt.ylabel("Bill Length (mm)")
plt.xlabel("Species")
plt.title("\nDistribution of Bill Length Area across Penguins Species\n", pad =10, fontsize=18)

# and save it
plt.savefig("penguins.png",dpi =500, format="png")
```
